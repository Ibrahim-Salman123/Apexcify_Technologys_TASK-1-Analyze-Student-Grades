## <span style=color:yellow;>****_Calculate Each Student's Average Score_****</span>

## <span style=color:yellow;>****_Student with the Highest Average_****</span>

```python
top_student = df.loc[df["Average"].idxmax()]

print("Top Student Details:")
print(top_student[["NAME OF THE STUDENT", "Math", "English", "Science", "Average"]])
```
